## 將姓名轉為頭銜

```python
all_df['Title'] = all_df.Name.str.extract(' ([A-Za-z]+)\.', expand = False)
```

```python
all_df.head()
```

```python
all_df['Title'].value_counts()
```

```python
all_df['Title2'] = all_df['Title'].replace(['Mlle','Mme','Ms','Dr','Major','Lady','Countess','Jonkheer','Col','Rev','Capt','Sir','Don','Dona'],
         ['Miss','Mrs','Miss','Mr','Mr','Mrs','Mrs','Mr','Mr','Mr','Mr','Mr','Mr','Mrs'])
```

```python
all_df['Title2'].value_counts()
```

```python
all_df = all_df.drop(['Ticket', 'Name'], axis=1)
```

```python
all_df.head()
```

```python
all_df['Fare'] = all_df['Fare'].fillna(all_df['Fare'].mean())
```

```python

all_df["Cabin"] = all_df['Cabin'].apply(lambda x : str(x)[0] if not pd.isnull(x) else 'NoCabin')
```

```python
all_df['Cabin'].value_counts()
```

```python
#all_df['Sex'] = all_df['Sex'].astype('category').cat.codes
```

```python
freq_port = train_df.Embarked.dropna().mode()[0]
```

```python
all_df['Embarked'] = all_df['Embarked'].fillna(freq_port)
```

```python
all_df.head()
```

```python
all_df = all_df.drop('Title', axis=1)
```

```python
all_df.head()
```

```python
all_df.info()
```

```python
median_group_age = all_df.groupby('Title2')['Age'].median()
```

```python
median_group_age
```

```python
for title in median_group_age.index:
    all_df.loc[(all_df.Age.isnull()) & (all_df.Title2 == title),'Age'] = median_group_age.loc[title]
```

```python
all_df.info()
```

```python
all_df = pd.get_dummies(all_df)
```

```python
all_df.head()
```

```python
scaler = StandardScaler()

all_df[['Age', 'Fare']] = scaler.fit_transform(all_df[['Age', 'Fare']].values)
#all_df['Fare'] = scaler.fit_transform(all_df['Fare'])
```

```python
all_df.head()
```

```python
# random forest
X = train_df.drop(labels=['Survived','PassengerId'],axis=1)
```

```python
X.head()
y = train_df['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.ensemble import VotingClassifier
```

# split the dataset 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,
                                                    random_state=0)
```

## Random forest

```python
rf = RandomForestClassifier(criterion='gini', 
                             n_estimators=1000,
                             min_samples_split=12,
                             min_samples_leaf=1,
                             oob_score=True,
                             random_state=1,
                             n_jobs=-1) 

rf.fit(X_train, y_train)
print("%.4f" % rf.oob_score_)
```

```python
scores = cross_val_score(rf, X_train, y_train, cv=5)
accuracy = scores.mean()
```

```python
print("Random forest Accuracy :", accuracy)
```

```python
rf.score(X_test, y_test)
```

## SVM

```python
svc_model = svm.SVC().fit(X_train, y_train)
```

```python
parameters = {'kernel': ['linear'], 'C': [0.1, 1, 10]}
svc_best_model = GridSearchCV(svc_model, parameters, cv=2, scoring='accuracy', return_train_score='cv_results_')
svc_best_model.fit(X_train, y_train)
```

```python
svc_best_model.score(X_test, y_test)
```

## Xgboost

```python
warnings.filterwarnings('ignore')
parameters= {'n_estimators':[100,200,500], 'learning_rate':[0.1,0.05,0.01,0.001], 'max_depth':[8,10,12]}
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
xgb_best_model = RandomizedSearchCV(xgb_model, parameters, cv=5, scoring='accuracy', return_train_score='cv_results_')
xgb_best_model.fit(X_train, y_train)
```

```python
xgb_best_model.score(X_test, y_test)
```

## Logistic regression

```python
parameters = {'penalty': ['l2'], 'max_iter': [100, 150,200],
                     'C': [1, 10, 100, 1000]}

logistic_model = LogisticRegression()
#logistic_model.fit(scale_all['StandardScaler']['x_train'], Y_train)
logistic_best_model = RandomizedSearchCV(logistic_model, parameters, cv=5, scoring='accuracy', return_train_score='cv_results_')
logistic_best_model.fit(X_train, y_train)

logistic_best_model.best_estimator_
```

```python
logistic_best_model.score(X_test, y_test)
```

## ensemble by voting

```python
vclf = VotingClassifier(estimators=[('svc', svc_best_model), ('rf', rf), ('xgb', xgb_best_model), ('lg', logistic_best_model)])
```

```python
vclf.fit(X_train, y_train)
```

```python
vclf.score(X_test, y_test)
```

```python
x_submit = test_df.drop(labels=['PassengerId'],axis=1)
pred = vclf.predict(x_submit)
```

```python
submit = pd.DataFrame({"PassengerId": test_df['PassengerId'],
                      "Survived":pred.astype(int)})
```
